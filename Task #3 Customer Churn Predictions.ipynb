```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/drive/MyDrive/Codsoft/Churn/Churn_Modelling.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.duplicated().sum()
```

```python
data.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
data = pd.get_dummies(data,columns=['Geography','Gender'],drop_first=True)
```

```python
data.head()
```

```python
X=data.drop(columns=['Exited'])
y=data['Exited']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_trf = scaler.fit_transform(X_train)
X_test_trf = scaler.transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(3,activation='relu',input_dim=11))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
his = model.fit(X_train_trf,y_train,epochs=42,validation_split=0.2)
```

```python
model.layers[0].get_weights()
```

```python
model.predict(X_test_trf)
```

```python
y_lg= model.predict(X_test_trf)
```

```python
y_pred=np.where(y_lg>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
